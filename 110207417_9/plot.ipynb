# 畫出一個 polygon (多邊形)，在頂點標示 (x,y) 坐標

````python
import matplotlib.pyplot as plt

xs = [1, 2, 4, 6, 5, 1]
ys = [2, 4, 5, 4, 3, 2]

plt.plot(xs, ys)

for x,y in zip(xs, ys):
    label = '({},{})'.format(x,y)
    
    plt.annotate(label, # this is the text
                 (x,y), # these are the coordinates to position the label
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='center') # horizontal alignment can be left, right or center
plt.ylim([0, 6])
plt.xlim([0, 7])
plt.show()
````

# 將 2 個子圖 (1 x 2) 並排

````python
import matplotlib.pyplot as plt

xs = [1, 2, 4, 6, 5, 1]
ys = [2, 4, 5, 4, 3, 2]

fig = plt.figure(figsize=(18, 6))

ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

ax1.plot(xs, ys)
ax1.set_ylim([0, 6])
ax1.set_xlim([0, 7])

for x,y in zip(xs, ys):
    label = '({},{})'.format(x,y)
    
    ax1.annotate(label, # this is the text
                 (x,y), # these are the coordinates to position the label
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='center') # horizontal alignment can be left, right or center
    
x2 = [1, 2]
y2 = [2, 4]

# Plot the x2 and y2 data points
ax1.plot(x2, y2)

# Fill the region with color
ax1.fill_between(x2, 0, y2, color='orange')

ax2.plot(xs, ys)
ax2.set_ylim([0, 6])
ax2.set_xlim([0, 7])

for x,y in zip(xs, ys):
    label = '({},{})'.format(x,y)
    
    ax2.annotate(label, # this is the text
                 (x,y), # these are the coordinates to position the label
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='center') # horizontal alignment can be left, right or center
    
x2 = [2, 4]
y2 = [4, 5]

# Plot the x2 and y2 data points
ax2.plot(x2, y2)

# Fill the region with color
ax2.fill_between(x2, 0, y2, color='orange')

plt.show()
````

### <font color="red">設計一個 function 執行畫子圖的步驟，請注意 function 的參數</font>

````python
import matplotlib.pyplot as plt

def plot_ax(ax, x_arr, y_arr):
    ax.plot(xs, ys)
    ax.set_ylim([0, 6])
    ax.set_xlim([0, 7])

    for x,y in zip(xs, ys):
        label = '({},{})'.format(x,y)

        ax.annotate(label, # this is the text
                     (x,y), # these are the coordinates to position the label
                     textcoords="offset points", # how to position the text
                     xytext=(0,10), # distance from text to points (x,y)
                     ha='center') # horizontal alignment can be left, right or center

    # Plot the x_arr and y_arr data points
    ax.plot(x_arr, y_arr)

    # Fill the region with color
    ax.fill_between(x_arr, 0, y_arr, color='orange')

xs = [1, 2, 4, 6, 5, 1]
ys = [2, 4, 5, 4, 3, 2]

fig = plt.figure(figsize=(15, 4))

ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)

xs1 = [1, 2]
ys1 = [2, 4]
xs2 = [2, 4]
ys2 = [4, 5]
xs3 = [4, 6]
ys3 = [5, 4]

plot_ax(ax1, xs1, ys1)
plot_ax(ax2, xs2, ys2)
plot_ax(ax3, xs3, ys3)

plt.show()
````

### <font color="red">設計一個 function 執行畫子圖的步驟，請注意 function 的參數</font>

````python
import matplotlib.pyplot as plt

def plot_ax(ax, x_arr, y_arr, color):
    ax.plot(xs, ys)
    ax.set_ylim([0, 6])
    ax.set_xlim([0, 7])

    for x,y in zip(xs, ys):
        label = '({},{})'.format(x,y)

        ax.annotate(label, # this is the text
                     (x,y), # these are the coordinates to position the label
                     textcoords="offset points", # how to position the text
                     xytext=(0,10), # distance from text to points (x,y)
                     ha='center') # horizontal alignment can be left, right or center

    # Plot the x_arr and y_arr data points
    ax.plot(x_arr, y_arr)

    # Fill the region with color
    ax.fill_between(x_arr, 0, y_arr, color=color)

xs = [1, 2, 4, 6, 5, 1]
ys = [2, 4, 5, 4, 3, 2]

fig = plt.figure(figsize=(15, 12))

ax1 = fig.add_subplot(331)
ax2 = fig.add_subplot(332)
ax3 = fig.add_subplot(333)
ax4 = fig.add_subplot(334)
ax5 = fig.add_subplot(335)
ax6 = fig.add_subplot(336)
ax7 = fig.add_subplot(337)

xs1 = [1, 2]
ys1 = [2, 4]
xs2 = [2, 4]
ys2 = [4, 5]
xs3 = [4, 6]
ys3 = [5, 4]
xs4 = [6, 5]
ys4 = [4, 3]
xs5 = [5, 1]
ys5 = [3, 2]
xs6 = [1, 2, 4, 6]
ys6 = [2, 4, 5, 4]
xs7 = [6, 5, 1]
ys7 = [4, 3, 2]

plot_ax(ax1, xs1, ys1, 'orange')
plot_ax(ax2, xs2, ys2, 'orange')
plot_ax(ax3, xs3, ys3, 'orange')
plot_ax(ax4, xs4, ys4, 'green')
plot_ax(ax5, xs5, ys5, 'green')
plot_ax(ax6, xs6, ys6, 'orange')
plot_ax(ax7, xs7, ys7, 'green')

plt.show()
````

# 計算 polygon (多邊形) 面積的步驟

以上圖為例，構成多邊形的頂點為：

```
P1 = (1,2)
P2 = (2,4)
P3 = (4,5)
P4 = (6,4)
P5 = (5,3)
```

將相鄰 2 點的線段與 X 軸構成梯形，則面積為： $(上底+下底) \times 高 \div 2$

例如：
$$
\frac{(P1_y + P2_y) \times (P2_x - P1_x)}{2}
$$


逐一計算各梯形的面積後加總即可得多邊形的面積

<font color="red">**注意：**按照點位的順序計算面積可能得到負值，須取絕對值</font>
